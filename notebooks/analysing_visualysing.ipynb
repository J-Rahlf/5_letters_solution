```python
import pandas as pd
import matplotlib.pyplot as plt

# Initialize an empty DataFrame to store the counts
letter_counts = pd.DataFrame()

# Loop through each column (except the first one) and count the occurrences
for column in preprocessed_df.columns[1:]:
    counts = preprocessed_df[column].value_counts()
    letter_counts[column] = counts

# Fill missing values with 0 and sum across columns to get the total counts
letter_counts.fillna(0, inplace=True)
letter_counts["Total"] = letter_counts.sum(axis=1)

# Plot a bar chart of the letter counts
plt.figure(figsize=(10, 6))
letter_counts["Total"].plot(kind="bar")
plt.title("Occurrences of Letters in Total")
plt.xlabel("Letter")
plt.ylabel("Total")
plt.show()
```

```python
import pandas as pd

# Laden der ursprünglichen CSV-Datei
preprocessed_words_df = pd.read_csv("../preprocessed_words.csv")

# Berechnung des Scores für jedes Wort
scores = []
for word in preprocessed_words_df["Word"]:
    score = 0
    for idx, letter in enumerate(word):
        letter_count = preprocessed_df[f"L{idx+1}"].value_counts().get(letter, 0)
        score += letter_count
    scores.append(score)

# Hinzufügen des Scores zum DataFrame
preprocessed_words_df["Score"] = scores

# Sortieren des DataFrames nach dem Score in absteigender Reihenfolge
preprocessed_words_df = preprocessed_words_df.sort_values(by="Score", ascending=False)

# Speichern des DataFrames mit dem Score in einer neuen CSV-Datei
preprocessed_words_df.to_csv("../preprocessed_words_with_score.csv", index=False)
```

```python
def get_input_word():
    while True:
        input_word = input("Bitte geben Sie ein Wort mit 5 Buchstaben ein: ")
        if len(input_word) == 5 and input_word.isalpha():
            return input_word
        else:
            print("Das eingegebene Wort muss genau 5 Buchstaben enthalten und nur Buchstaben enthalten. Bitte versuchen Sie es erneut.")

def assign_statuses(input_word):
    statuses = []
    for i in range(5):
        while True:
            status = input(f"Geben Sie den Status für den Buchstaben '{input_word[i]}' (0, 1 oder 2) ein: ")
            if status in ['0', '1', '2']:
                statuses.append(int(status))
                break
            else:
                print("Ungültige Eingabe. Bitte geben Sie 0, 1 oder 2 ein.")
    return statuses

def main():
    global input_word, statuses
    
    input_word = get_input_word()
    statuses = assign_statuses(input_word)
    print("Eingegebenes Wort:", input_word)
    print("Status der Buchstaben:", statuses)



if __name__ == "__main__":
    main()
```

```python
import pandas as pd

print("Eingegebenes Wort:", input_word)
print("Status der Buchstaben:", statuses)




def check_word(word, input_word, statuses):
    # Initialize the result
    valid_word = True
    
    # Check if the word meets the criteria
    for i, (letter, status) in enumerate(zip(input_word, statuses)):
        
        # If the status is '2', check if the letter is at the correct position
        if status == 2 and input_word[i] != word[i]:
            valid_word = False
            break
        

        # Check if the status of the letter is correct
        elif status == 0:
            if letter in word:
                if letter in input_word:
                    valid_word = False
                    break  
         
        else:
            if status == 1 and ((letter not in word) or (letter in word and input_word[i] == word[i])):
                valid_word = False
                break     

    return valid_word

# Laden der vorverarbeiteten Wörter und ihrer Statuswerte aus der CSV-Datei
preprocessed_df = pd.read_csv('../preprocessed_words_with_score.csv')

# Initialisierung eines leeren DataFrames für die gefilterten Wörter
filtered_df = pd.DataFrame(columns=preprocessed_df.columns)

# Iteriere durch jeden Eintrag im DataFrame und überprüfe, ob das Wort den Kriterien entspricht
for index, row in preprocessed_df.iterrows():
    word = row['Word']
    # Überprüfe, ob das Wort den Kriterien entspricht
    if check_word(word, input_word, statuses):
        # Füge das Wort zum gefilterten DataFrame hinzu
        filtered_df = filtered_df.append(row, ignore_index=True)





# Anzeigen des gefilterten DataFrames
print(filtered_df.head(50))

```
